La segunda forma, es la que trabajaremos más pues nos da más herramientas, se basa en objetos a los cuales podremos aplicar funciones y acciones. Empezaremos parecido a la primera forma, pero ahora crearemos un objeto figura en una variable que llamaremos fig y en ella crearemos ejes para plotear

```python
%matplotlib inline
```

```python
import matplotlib
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-1,5,21)
x
```

```python
y = x**2
y
```

```python
fig = plt.figure()

axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])   # los dos primeros nos indicaran la posición de la intersección de ejes y 
                                            # Los otros dos son la proporción al tamaño de ancho y alto
axes.plot(x, y, 'r')

axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('Mi primera figura');
```

Ahora que tenemos el control de donde se crean los ejes, podemos crear más de uno:

* Crearemos una figura y dentro de esta otra figura más pequeña

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # serán los ejes principales
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3]) # serán los ejes de la figura más pequeña a dentro


# Figura principal
axes1.plot(x, y, 'r')
axes1.set_xlabel('x')
axes1.set_ylabel('y')
axes1.set_title('x al cuadrado')

# Figura insertada
axes2.plot(y, x, 'g')
axes2.set_xlabel('y')
axes2.set_ylabel('x')
axes2.set_title('raiz cuadrada de y');
```

En matplotlib existen funciones que ordenan nuestros ejes, una de esas es ``subplots()`` y se usa de la siguiente manera:

```python
fig, axes = plt.subplots()

axes.plot(x, y, 'r')
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('Figura');
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2) #Con esto nos generará una cuadricula de la forma 1x2 de figuras

for ax in axes:
    ax.plot(x, y, 'r')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_title('Figura')
fig.tight_layout() #Ajustará las posiciones para que no hayan superposiciones
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2) #Con esto nos generará una cuadricula de la forma 1x2 de figuras

axes[0].plot(x, y, 'b*-') 
axes[0].set_xlabel('x')
axes[0].set_ylabel('y')
axes[0].set_title('cuadrado')
axes[1].plot(y, x, 'y--') 
axes[1].set_xlabel('y')
axes[1].set_ylabel('x')
axes[1].set_title('Raiz')

fig.tight_layout() #Ajustará las posiciones para que no hayan superposiciones
```

```python
# histograma
n = np.random.randn(100000)
fig, axes = plt.subplots(1, 2, figsize=(12,4))

axes[0].hist(n)
axes[0].set_title("histograma")
axes[0].set_xlim((min(n), max(n)))

axes[1].hist(n, cumulative=True, bins=50)
axes[1].set_title("histograma acumulado")
axes[1].set_xlim((min(n), max(n)));
```

# 3. Seaborn

El principal problema de matplotlib es lidear con los dataframes de Pandas, por lo que Seaborn es una solución a ello, pues nos permitirá trabajar con dataframes de manera suave, nos dará estilos de ploteo e incluyen funciones estadisticas muy usadas

```python
rng = np.random.RandomState(0)
x = np.linspace(0, 10, 500)
y = np.cumsum(rng.randn(500, 6), 0)
```

```python
plt.plot(x, y)
plt.legend('ABCDEF', ncol=2, loc='upper left');
```

```python
import seaborn as sns
sns.set()
```

```python
plt.plot(x, y)
plt.legend('ABCDEF', ncol=2, loc='upper left');
```

La idea principal de usar Seaborn es que nos permite crear ploteos de data estadística con muchos comandos avanzados

En visualización de Data estadística, veremos muy seguido ploteos de histogramas para ver las distribuciones de las variables, en matplotlib veremos:

```python
import pandas as pd
```

```python
data = np.random.multivariate_normal([0, 0], [[5, 2], [2, 2]], size=2000)
data = pd.DataFrame(data, columns=['x', 'y'])

for col in 'xy':
    plt.hist(data[col],density = 1, alpha=0.5)
```

Mientras que Seaborn nos ofrece una función que usa una estimacion de densidad del kernel, con el que podremos ver las distribuciones de una menera más sencilla usando `` sns.kdeplot``

```python
for col in 'xy':
    sns.kdeplot(data[col], shade=True)
```

Y podemos ver estos dos juntos usando ``distplot ``

```python
sns.distplot(data['x'])
sns.distplot(data['y']);
```

Podremos visualizar la data en dos dimensiones usando toda la data pero usando la funcion ``jointplot`` podremos ver la distribución conjunta y las respectivas distribuciones marginales

```python
with sns.axes_style('white'):
    sns.jointplot("x", "y", data, kind='kde');
```

```python
# Usando el estilo hexagonal
with sns.axes_style('white'):
    sns.jointplot("x", "y", data, kind='hex')
```

Si generalizamos esto para más de 2 dimensiones podremos generar estos ploteos de par en par. Esto se usa mucho para ver correlaciones entre data de muchas dimensiones, veremos esto con la base de datos 'iris' que enumera las medidas de pétalos y sépalos de tres especies de iris (flor):

```python
iris = sns.load_dataset("iris")
iris.head()
```

```python
sns.pairplot(iris, hue='species', height=2.5);
```
